```python
import cv2
import glob
from YoloV5EngineDetector import Detector
```

```python
detector = Detector()
detector.load()
```

```python
from matplotlib import pyplot as plt

rootDir = "../../examples/images/*"
for img_path in glob.glob(rootDir):
    img = cv2.imread(img_path)
    print(img_path, img.shape)
    targetBoxes = detector.detect_bbox([img])
    print("targetBoxes", targetBoxes)

    for targetBox in targetBoxes:
        cv2.rectangle(img, 
                      (int(targetBox[0]), int(targetBox[1])), 
                      (int(targetBox[2]), int(targetBox[3])), 
                      (0,120,255), 
                      3)
    plt.imshow(img)
    plt.show()
```
